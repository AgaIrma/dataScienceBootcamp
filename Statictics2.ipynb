```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
#
# Draw random samples from normal distribution
# with a predefined mean and standard deviation
#
x = np.random.normal(loc=3, scale = 1, size=100)
#
# Plot probability distribution function
#
x_sorted = np.sort(x)
plt.figure(figsize=(7, 5))
plt.plot(x_sorted, norm.pdf(x_sorted, loc=3, scale=1))
plt.title("Normal distribution", fontsize=16)
plt.show()

df = pd.DataFrame(x, columns=list('A'))
print(df)

#obliczenie średniej, kwartyli czy wykresów pudełkowych
values = df.loc[df['A'].notnull(), 'A'].values
print(values)

#histogram
plt.hist(values, bins=20)
plt.show()


#srednia
mean = np.round(np.mean(values))
print(f'mean value {mean}')

#histogram ze srednia skad wiedziec na ile kubelkow podzielic
plt.hist(values, bins=20)
plt.axvline(x = mean, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.show()

#qwartyle
q0 = np.quantile(values, 0.0)
q1 = np.round(np.quantile(values, 0.25))
median = np.round(np.quantile(values, 0.5))
q3 = np.round(np.quantile(values, 0.75))
q4 = np.round(np.quantile(values, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')


```
